```python
#Libraries we'll need 
from collections import Counter
from sklearn.linear_model import LogisticRegression
from nltk.tokenize import word_tokenize
from sklearn import datasets, neighbors, linear_model
from sklearn.svm import SVC
import scipy.stats
from sklearn.feature_extraction import DictVectorizer
import numpy as np 
import pandas as pd
import string
import re
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from collections import defaultdict
import operator
import csv
```

```python
#initially reading in data
df = pd.read_csv('articles_train_degendered_filtered.csv',encoding='utf8')
```

```python
#selecting features and labels 
data = df[['sentence','gender']]
```

```python
#preprocessing: eliminating punctuation
#print("Cleaning data...")
def remove_punctuations(text):
    for punctuation in string.punctuation:
        #text = str(text).encode('utf-8').replace(punctuation, ' ')
        text = text.replace(punctuation, ' ')
    #text = re.sub( '\s+', ' ', text ).encode('utf-8').strip()
    text = re.sub( '\s+', ' ', text ).strip()
    return text
```

```python
data.loc[:, 'sentence'] = data.loc[:, 'sentence'].apply(remove_punctuations)
```

```python
print(data.shape)
```

```python
# returns list of sorted unigrams and bigrams
def sort_uni_bigrams():
    all_unigrams = defaultdict(int)
    for index, row in data.iterrows():
        text = row['sentence']
        words = word_tokenize(text)
        for i in range(1, len(words)):
            uni = words[i]
            all_unigrams[uni] += 1
    sorted_unigrams = sorted(all_unigrams.items(), key=operator.itemgetter(1), reverse=True)
    return sorted_unigrams
```

```python
def sort_bigrams():
    all_bigrams = defaultdict(int)
    for index, row in data.iterrows():
        text = row['sentence']
        words = ['<S>'] + word_tokenize(text) + ['</S>']
        for i in range(1, len(words)):
            bi = words[i-1] + " " + words[i]
            all_bigrams[bi] += 1
    sorted_bigrams = sorted(all_bigrams.items(), key=operator.itemgetter(1), reverse=True)
    return sorted_bigrams
```

```python
sorted_unigrams, sorted_bigrams = sort_uni_bigrams()
```

```python
#top 5k unigrams and top 10k bigrams for the case where unigrams and bigrams are being used together

top_unigrams = {uni[0]: uni[1] for uni in sorted_unigrams[:5000]}
top_bigrams = {bi[0]: bi[1] for bi in sorted_bigrams[:10000]}
```

```python
# given sentence, and list of top unigrams and bigrams
# returns feature dicts for unigrams and bigrams
def uni_bigrams(text):
    words = ['<S>'] + word_tokenize(text) + ['</S>']
    unigrams = defaultdict(int)
    bigrams = defaultdict(int)
    for i in range(1, len(words)):
        uni = words[i]
        bi = words[i-1] + " " + words[i]
        if uni in top_unigrams:
            unigrams[uni] += 1
        else:
            unigrams['UNK'] += 1
        if bi in top_bigrams:
            bigrams[bi] += 1
        else:
            bigrams['UNK UNK'] += 1
    return (unigrams, bigrams)
```

```python
#converting samples into features: top 5k unigrams + top 10k bigrams
feat_dicts = []
labels = []
for index, row in data.iterrows():
    text = row['sentence']
    label = row['gender']
    unigrams, bigrams = uni_bigrams(text)
    feature_dict = {**unigrams, **bigrams}
    feat_dicts.append(feature_dict)
    labels.append(label)
```

```python
vectorizer = DictVectorizer(sparse=True)
features = vectorizer.fit_transform(feat_dicts)
```

```python
#Model: Logistic Regression right out of the box 
logistic = linear_model.LogisticRegression(verbose=3)
logistic_model = logistic.fit(features, labels)
```

```python
df_dev = pd.read_csv('articles_dev_degendered_filtered.csv',encoding='utf8')
data_dev = df_dev[['sentence','gender']]
data_dev.loc[:, 'sentence'] = data.loc[:, 'sentence'].apply(remove_punctuations)
```

```python
feat_dicts_dev = []
labels_dev = []
for index, row in data_dev.iterrows():
    text = row['sentence']
    label = row['gender']
    unigrams, bigrams = uni_bigrams(text)
    feature_dict = {**unigrams, **bigrams}
    feat_dicts_dev.append(feature_dict)
    labels_dev.append(label)
```

```python
vectorizer_dev = DictVectorizer(sparse=True)
features_dev = vectorizer_dev.fit_transform(feat_dicts_dev)
```

```python
logistic_predictions = logistic.predict(features_dev)
```

```python
print(classification_report(labels_dev, logistic_predictions))
```
